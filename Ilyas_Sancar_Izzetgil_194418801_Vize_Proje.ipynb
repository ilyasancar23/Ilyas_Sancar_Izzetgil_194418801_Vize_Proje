***3.ADIM:*** Tüm bağımsız değişkenlerimizi içeren Özellikler Matrisi oluşturulur.

(Burada ".iloc" ile sayısal değer belirterek işlem yapabiliriz, buna alternatif olarak ".drop" ile belirlenen string ile de özellikler matrisini oluşturabiliriz)

```python
#X = data.iloc[:,0:-1].values
X = data.drop(["Outcome"], axis=1)
print(X)
```

***4.ADIM:***Bağımlı değişken değerlerimizi içerecek bir bağımlı değişken vektörü oluştururuz

(Yukarıdaki ile aynı şekilde ister ".iloc" ile ister ".drop" ile bağımlı değişken vektörü oluşturabiliriz)

```python
#y = data.iloc[:,-1].values
y = data["Outcome"]
print(y)
```

***5.ADIM:*** Veri setimizi, eğitim aşamasında verinin yüzde 80'i oranda olacak ve verinin yüzde 20'si test aşamasında olacak şekilde eğitim ve test veri setlerine ayırırız

"random_state", datayı her seferinde farklı yerlerinden böler. Belirlenen random_state değeri, her seferinde o değere göre böler

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
#from sklearn.model_selection import train_test_split
#X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

***6.ADIM:*** şırı yüksek değerli değişkenlerin, aşırı düşük değerli değişkenlere hakim olma ve düşük değerli değişkenlerin model tarafından ihmal edilme olasılığı için tüm değişkenlerin aynı ölçeğe dönüştürülmesi gerekir. Dağılımdan bağımsız olarak herhangi bir veri kümesi için kullanılabilen evrensel bir teknik olan Standardizasyon kullanılır

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

***7.ADIM:***  "Sequential" Keras sınıfı kullanılarak YSA başlatılır

```python
ann = tf.keras.models.Sequential()
```

***8.ADIM:*** En uygun değer olan 6'yı alan "units" ve hangi aktivasyon fonksiyonunun kullanılacağını belirten "activation" girişleri tanımlanır

```python
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
```

***9.ADIM:*** İki gizli katman oluşturulacağından aynı adım tekrarlanır

```python
ann.add(tf.keras.layers.Dense(units=6,activation="relu"))
```

***10.ADIM:*** Devamındaki işlemlerde Dense sınıfı kullanılarak, çıktı olarak("Potability") sadece iki sınıfa (1 ve 0) sahip olacağımız bir ikili sınıflandırma yapacak bir nöron tahsis edilir ve ikili sınıflandırma için sigmoid aktivasyon fonksiyonu kullanılır

```python
ann.add(tf.keras.layers.Dense(units=1,activation="sigmoid"))
```

***11.ADIM:*** Deney sonuçları için kritik öneme sahip "optimizer", "loss", "metrics" girdileri ayarlanır ve YSA derlenir

```python
ann.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=128,epochs=1000)
```
